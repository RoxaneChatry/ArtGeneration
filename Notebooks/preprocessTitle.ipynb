```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from skimage.io import imread
import pandas as pd
# from IPython.display import Image, display

file_path = "/content/drive/Shareddrives/PFE artists/data/base-joconde-extrait.csv"
```

```python
data = pd.read_csv(file_path, sep=";", quotechar="\"")
data.describe()
```

```python
print(data.columns)
data.head()
```

```python
# liste des domaines des oeuvres
data["Domaine"].describe()
```

```python
liste_domaines = pd.unique(data["Domaine"])
print(liste_domaines)
all_domaines = []
for element in liste_domaines:
  all_elements = str(element).split(";")
  for el in all_elements:
    if not (el in all_domaines):
      all_domaines.append(el)
print(all_domaines)
print("il y a {} domaines différents".format(len(all_domaines)))
```

```python
# on ne récupère que les domaines qui nous intéressent : peinture et dessin
# or chaque oeuvre peut avoir plusieurs domaines 
data_peintures = data[(data.Domaine.str.contains("peinture")) | (data.Domaine.str.contains("dessin"))]
data_peintures.Domaine.describe()
```

```python
# On s'intéresse aux titres des oeuvres sélectionnées
data_peintures.Titre.describe()
```

```python
# Les objets "(Sans titre)" ne nous intéressent pas
# data_sauf_nan = data_peintures[data_peintures["Titre"].isnull() == False]
data_titrees = data_peintures[(data_peintures.Titre.str.contains("Sans titre") == False) & (data_peintures["Titre"].isnull() == False)]
data_titrees.describe()
```

```python
liste_titres= data_titrees.Titre.replace('"',"").astype('str')
```

```python
# Liste des titres
liste_titres
```

```python
df = pd.DataFrame(liste_titres)
df.to_csv("/content/drive/Shareddrives/PFE artists/title_generator/title_data.txt"  ,index = None)
```
